## LLM: Large Language Model

A languaje model is a model that predict the next token based on the tokens you have so far.And example would be using words as tokens and predicting "how are you?" as "how - " --> "how are -" --> "how are you?"

An LLM is a model with millions of parameters trained on billion of data, but that work as a languaje model, generally speaking.The base architechture for LLM's is transformers, creating encoder-decoder or just decoder neural neworks.





## RAG: Retrieval Agumented Generation

Retrieval = Search
Generation = LLM

This kind of systems consist of using a serach engine to pass information as context to an LLM that we use to generate answers.
